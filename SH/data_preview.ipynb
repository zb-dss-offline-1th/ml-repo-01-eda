```python
data2 = pd.read_csv('./지하철/서울교통공사_역별 일별 시간대별 승하차인원 정보_20220531.csv', encoding='cp949')
data2
```

```python
data2.drop('연번', axis=1, inplace=True)
```

```python
data2.info()
```

```python
data2.columns
```

```python
data2.drop(['12시-13시', '13시-14시',
       '14시-15시', '15시-16시', '16시-17시', '17시-18시', '18시-19시', '19시-20시',
       '20시-21시', '21시-22시', '22시-23시', '23시 이후'], axis=1, inplace=True)
```

```python
data2.head()
```

```python
data2.drop('역번호', axis=1, inplace=True)
```

```python
data2['날짜'] = pd.to_datetime(data2['날짜'], format='%Y-%m-%d')
data2.head()
```

```python
data2.info()
```

```python
data2['Weekday'] = 0
```

```python
for idx, row in data2.iterrows():
    if row['날짜'].weekday() > 4:
        data2.loc[idx, 'Weekday'] = 1
```

```python
data2
```

```python
df = data2.loc[(data2['역명']== '삼각지') | (data2['역명']== '이태원') | (data2['역명']== '신용산')]
```

```python
df.head()
```

```python
df = df[df['구분'] == '하차']
df
```

```python
df = df[df['Weekday'] == 0]
```

```python
df
```

```python
df['Post'] = 0
```

```python
df.head()
```

```python
threshold = '2022-05-10'
thresh = pd.to_datetime(threshold, format='%Y-%m-%d')
for idx, row in df.iterrows():
    if row['날짜'] > thresh:
        df.loc[idx, 'Post'] = 1
```

```python
df.head()
```
